Implemente a regressão linear simples para prever o preço de casas usando o conjunto de dados Boston Housing. Avalie o modelo utilizando o R² e o erro médio absoluto (MAE).

```python
# Importado as bibliotecas necessárias.:
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
# Importado o arquivo csv.
dataset = pd.read_csv("/content/BostonHousing.csv", encoding="utf-8")
```

```python
# Exibindo as 10 primeiras linhas do dataset
dataset.head(10)
```

```python
# Realizando a limpeza de valores nulos:
dataset = dataset.dropna(subset=['rm'])
```

```python
# Separando os dados de treinamento e teste.
feature, classe = dataset[['rm']], dataset['medv']
```

```python
# Separando os conjuntos de dados para treinamento e validação.
feature_treinamento, feature_validacao, classe_treinamento, classe_validacao = train_test_split(feature, classe, test_size=0.30)
```

```python
# Criando o modelo de Regressão Linear.
modelo_RL = LinearRegression()
modelo_RL.fit(feature, classe)
```

```python
# Fazendo a predição com os valores de validação.
classe_predicao = modelo_RL.predict(feature_validacao)
```

```python
# Métrica MAE.
print(f"Média Absoluta de Erro: {mean_absolute_error(classe_validacao, classe_predicao):.2f}")
```

```python
# Métrica R2.
print(f"R² Score: {r2_score(classe_validacao, classe_predicao):.2f}")
```
