# Classificação com K-Nearest Neighbors (KNN)
Construa um modelo KNN para prever a classe de um conjunto de dados (por exemplo, Wine). Utilize validação cruzada para ajustar o número de vizinhos e avalie o desempenho com a métrica F1-Score.

```python
# Importado as bibliotecas necessárias.
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
# Importado o arquivo csv.
dataset = pd.read_csv("/content/wine_dataset.csv", encoding="utf-8")
```

```python
# Exibindo as 10 primeiras linhas do dataset.
dataset.head(10)
```

```python
# Transformado a coluna species em dados categóricos.
dataset['style'] = dataset['style'].replace({'white': 0, 'red': 1}).infer_objects()
```

```python
# Definindo as features do modelo.
features, classe = dataset.drop(['style'], axis = 1), dataset['style']
```

```python
# Definindo a classe do modelo.
x_treino, x_val, y_treino, y_val = train_test_split(features, classe, test_size=0.30)
```

```python
# Realizando treinando do modelo KNN.
modelo_KNN = KNeighborsClassifier()
modelo_KNN.fit(x_treino, y_treino)
```

```python
# Predição do conjunto de treinamento.
y_pred_treino = modelo_KNN.predict(x_val)
```

```python
# Utilização da métrica do F1 Score:
print(f"F1-Score do Modelo: {f1_score(y_val, y_pred_treino)}")
```

```python
# Realizando a validação cruzada dos dados.
scores = cross_val_score(modelo_KNN, features, classe, cv=5, scoring='accuracy')
```

```python
# Exibido a Acurácia.
print("Acurácia da Validação Cruzada: %0.2f" % (scores.mean()))
```
