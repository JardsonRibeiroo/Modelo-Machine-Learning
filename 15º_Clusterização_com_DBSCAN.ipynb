```python
# Importado as bibliotecas necessárias.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
# Importado o arquivo csv.
dataset = pd.read_csv("/content/vendas.csv", encoding="utf-8")
```

```python
# Exibindo as 10 primeiras linhas do dataset.
dataset.head(10)
```

```python
dataset = dataset.drop(columns=['LOJA', 'DATA_VENDAS','CLIENTE',])
```

```python
# Exibindo as 10 primeiras linhas do dataset.
dataset.head(10)
```

```python
# Selecionar as colunas relevantes para a clusterização
X = dataset[['VALOR', 'MES', 'ANO', 'ID_PRODUTO']]
```

```python
# Inicializar e treinar o modelo DBSCAN
dbscan = DBSCAN(eps=500, min_samples=5)
labels = dbscan.fit_predict(X)
```

```python
# Plotar os clusters
plt.scatter(X['VALOR'], X['MES'], c=labels, cmap='viridis', s=50)
plt.title('Clusters DBSCAN')
plt.xlabel('Valoe')
plt.ylabel('Mês')
plt.show
```
